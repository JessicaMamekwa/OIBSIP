```python
#Exploring Numerical columns 
car_data.hist(bins=20, figsize=(25, 20))
plt.show()
```

```python
plt.figure(figsize=(18, 8))
plt.subplot(2, 4, 2)
sns.scatterplot(x ="Year", y ="Selling_Price", data = car_data) #Finding relationship between the Year and selling price
plt.subplot(2, 4, 3)
sns.scatterplot(x = "Present_Price", y="Selling_Price", data = car_data) #Finding relationship between the Present Price and selling price
plt.subplot(2, 4, 4)
sns.scatterplot(x ="Driven_kms", y = "Selling_Price", data = car_data) #Finding relationship between the driven kilometers and selling price
plt.tight_layout()
plt.show()
```

```python
#Finding Relationship between Cars and it's Selling price using the Boxplot
plt.figure(figsize=(14,10))
sns.boxplot(x="Car_Name", y="Selling_Price", data= car_data)
plt.xticks(rotation=90)  
plt.show()
```

```python
plt.figure(figsize=(16, 8))
plt.subplot(1, 2, 1)
sns.boxplot(x="Fuel_Type", y="Selling_Price", data= car_data)
plt.title("Relation between Fuel type and the Selling Price")
plt.subplot(1, 2, 2)
sns.boxplot(x="Transmission", y="Selling_Price", data=car_data)
plt.title("Relation between Transmission and Selling price")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (12,7))
ax=sns.swarmplot(x="Age",y="Selling_Price",data=car_data, palette = "winter")
ax.set_xticklabels(ax.get_xticklabels(),rotation=40,ha='right')
plt.title('Relation Between Age and Selling Price')
plt.show()
```

```python
plt.tight_layout(pad = 1)
sns.pairplot(car_data, hue = "Selling_Price", height = 4, aspect = 1.2, palette = "coolwarm")
```
